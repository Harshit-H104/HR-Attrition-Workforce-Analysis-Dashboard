```python
# Check Class Imbalance
df['attrition'].value_counts()
```

```python
# Attrition by Department
df.groupby('department')['attrition'].value_counts(normalize=True) * 100
```

```python
# Attrition by Gender
df.groupby('gender')['attrition'].value_counts()
```

```python
# Attrition by Overtime
df.groupby('overtime')['attrition'].value_counts()
```

```python
# Attrition vs MonthlyIncome 
df.groupby('attrition')['monthlyincome'].mean()
```

```python
#Attrition vs JobSatisfaction
df.groupby('jobsatisfaction')['attrition'].value_counts()
```
